```python
import torch as t
x = t.tensor(2.0, requires_grad=True)
z = t.tensor(1.0, requires_grad=False)
```

```python
y = 2*x
t = y * z
t.backward()
```

```python
print(x.grad, y.grad, z.grad, t.grad)
```

```python
class Gs:
    def __init__(self, val = 1):
        self.val = val
    def get_value(self):
        self.val = self.val + 1
        return self.val - 1
```

```python
train_step = Gs()
x = train_step.get_value()
y = train_step.val
print(x, y)
```
